A root K/lateral ft (md1/
2 *ft) is used as the output feature in this exercise. A root K is a productivity metric obtained from rate transient analysis (RTA) in unconventional reservoirs and it is equivalent to kh in conventional reservoirs. 
A root K is simply the cross-sectional area
multiplied by the square root of permeability. It is essentially a productivity metric that shows the strength of each well.

```python
# drop TOC and matrix perm since high correlation with porosity

df1=df.drop(['TOC (%)','Matrix Perm (nd)' ], axis=1)
```

```python
df1.head()
```

```python
#normalizing the data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
scale=scaler.fit_transform(df1)
```

```python
df1_scaled=pd.DataFrame(scale, columns=['Porosity (%)',
'Acoustic impedance (kg/m2s*106)', 'Brittleness Ratio',
'Vitrinite Reflectance (%)', 'Aroot(K)'])
```

```python
df1_scaled.head()
```

```python
#take xscaled as input features and the yscaled as output
y_scaled=df1_scaled['Aroot(K)']
x_scaled=df1_scaled.drop(['Aroot(K)'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaled, y_scaled, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
coef_df=pd.DataFrame(lm.coef_, x_scaled.columns, columns=['Coefficients'])
```

```python
coef_df
```

```python
lm.intercept_   #(this is b)
```

```python
y_pred=lm.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('actual test')
plt.ylabel('predicted test')
```

```python
#we have to compare the y_prediction values with the original values because we have to calculate the accuracy of our model, which was implemented by a concept called r2_score

from sklearn.metrics import r2_score
test_set_r2=r2_score(y_test, y_pred)
```

```python
round(test_set_r2, 2)
#accuracy score between actual and predicted 
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error
```

```python
MAE=mean_absolute_error(y_test, y_pred)
MSE=mean_squared_error(y_test, y_pred)
RMSE=np.sqrt(MSE)
print(f'MAE:{np.round(MAE, 4)} \n\nMSE:{np.round(MSE, 4)} \n\nRMSE:{np.round(RMSE,4)}')
```
