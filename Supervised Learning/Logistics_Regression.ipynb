```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(),annot=True, cmap='gnuplot2_r')
#below we can see that there is'nt high correlations so no need to drop features 
```

```python
#Quitting is the output feature we need 
y=df['Quitting']
x=df.drop(['Quitting'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(penalty='l2',C=1, solver='saga')
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True,
cmap='viridis')
```

```python
print(classification_report(y_test,y_pred))
```

```python
#to conclude, 115 out of 122 was correctly classified in class 0 and 7 as uncorrect
#123 out of 128 in class 1 5 as uncorrect 
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
#we can use this to see the accuracy also
```
