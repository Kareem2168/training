```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv')
df
```

```python
df.drop(['longitude'], axis='columns', inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
x=df.drop(['median_house_value'], axis='columns')
y=df['median_house_value']
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
scaler=MinMaxScaler()
knn=KNeighborsRegressor()
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
from sklearn.linear_model import LinearRegression
```

```python
pipe=Pipeline([('scaler', StandardScaler()), 
               ('regressor',LinearRegression())])

grid_param=[
            {'scaler':[StandardScaler(),scaler],
             'regressor':[knn], 
             'regressor__n_neighbors':[2],
             'regressor__metric':['euclidean', 'minkowski']},
            
            {'scaler':[StandardScaler(),scaler],
             
             'regressor':[LinearRegression()]}
]


#############
gridsearch = RandomizedSearchCV(pipe, grid_param, cv=10, n_iter=3) # Fit grid search
best_model = gridsearch.fit(x_train,y_train)

###################
print('best score:',best_model.score(x_train, y_train))


print('\n\n\nbest estimator:',best_model.best_estimator_)

print('\n\n\nbest params:', best_model.best_params_)
```
